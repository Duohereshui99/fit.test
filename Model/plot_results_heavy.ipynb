```python
import numpy as np
import matplotlib.pyplot as plt


Sys=['288Fl','290Lv','294Og']
Z=[114,116,118]
T_half_expt=[0.64,8.3e-3,0.58e-3]
T_half_calc=[0.6387798626895655,0.006536245856827524,0.0005716369120320206]
T_half_2sigma=[0.2554177415374069,0.0052443584595577,0.0005214614134383]


# 计算比值和误差
ratios = [calc/expt for calc, expt in zip(T_half_calc, T_half_expt)]
ratio_errors = [2*sigma/expt for sigma, expt in zip(T_half_2sigma, T_half_expt)]

# 创建图形
plt.figure(figsize=(12, 6))

# 绘制比值点
for z, ratio, sys in zip(Z, ratios, Sys):
    plt.plot(z, ratio, 'o', markersize=8, label=f'{sys}')

# 绘制误差棒
plt.errorbar(Z, ratios, yerr=ratio_errors, fmt='none', 
             ecolor='black', elinewidth=2, capsize=5)

# 添加参考线
plt.axhline(y=1, color='gray', linestyle='--', linewidth=1)

# 设置坐标轴
plt.xticks(Z, [f"{z}({sys})" for z, sys in zip(Z, Sys)], fontsize=10)
plt.xlabel('Z', fontsize=12)
plt.ylabel('Theoretical/Experimental Ratio', fontsize=12)
#plt.ylim(0.77,1.25)  # 自动调整y轴范围

# 添加图例和网格
plt.legend(fontsize=10, loc='upper left')
plt.grid(True, which="both", ls="--", alpha=0.4)

plt.tight_layout()
plt.show()
```
